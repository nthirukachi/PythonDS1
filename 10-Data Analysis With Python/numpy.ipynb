```python
data = [
    [11,12,13,14],
    [21,22,23,24],
    [31,32,33,34],
    [41,42,43,44]
]

arr = np.array(data)
print(arr)
```

```python
arr.mean(axis=0)  # Mean of each column
print(arr.mean(axis=0))
arr.mean(axis=1)  # Mean of each row
print(arr.mean(axis=1))
arr.sum(axis=0)  # Sum of each column
print(arr.sum(axis=0))
arr.sum(axis=1)  # Sum of each row
print(arr.sum(axis=1))
```

```python
import numpy as np

arr = np.random.rand(3, 4)
print("Original Array:\n", arr)

squared_arr = arr ** 2
print("\nSquared Array:\n", squared_arr)

greater_than_0_5 = squared_arr[squared_arr > 0.5]
print("\nElements greater than 0.5:\n", greater_than_0_5)
```
