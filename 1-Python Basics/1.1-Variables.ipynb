### Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python

print(age, type(age))
print(height, type(height))
# Type Conversion Examples
# Integer to float
age_float = float(age)
print("Integer to float:", age_float, type(age_float))

# Integer to string
age_str = str(age)
print("Integer to string:", age_str, type(age_str))

# Integer to boolean
age_bool = bool(age)
print("Integer to boolean:", age_bool, type(age_bool))

# Float to integer
height_int = int(height)
print("Float to integer:", height_int, type(height_int))

# Float to string
height_str = str(height)
print("Float to string:", height_str, type(height_str))

# Float to boolean
height_bool = bool(height)
print("Float to boolean:", height_bool, type(height_bool))

# Boolean to integer
student_int = int(is_student)
print("Boolean to integer:", student_int, type(student_int))

# Boolean to float
student_float = float(is_student)
print("Boolean to float:", student_float, type(student_float))

# Boolean to string
student_str = str(is_student)
print("Boolean to string:", student_str, type(student_str))

# String to integer (if possible)
myname_int = None
try:
    myname_int = int(myname)
except ValueError:
    print(f"Cannot convert '{myname}' to int")

# String to float (if possible)
myname_float = None
try:
    myname_float = float(myname)
except ValueError:
    print(f"Cannot convert '{myname}' to float")

# String to boolean
myname_bool = bool(myname)
print("String to boolean:", myname_bool, type(myname_bool))

# String to list
name_list = list(name)
print("String to list:", name_list, type(name_list))
```
