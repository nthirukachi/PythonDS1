## ZIP

````python
names = ["Ramesh", "Raju", "Ramu", "Somu"]
scores = [80, 70, 65, 55]

for name, score in zip(names, scores):
  print(f"{name} scored {score}")
````

zip?

Clean way to iterate over two or more related lists.

Often used in:
- Working with rows and columns.
- Processing test cases (input/ output)
- Matching data from different sources (like names & email)

````python
students = ["Ramesh", "Raju", "Ramu", "Somu"]
scores = [80, 70, 65, 55]
subjects = ['Maths', "Science", "English", "Lang"]

for student, subject, score in zip(students, subjects, scores):
  print(f"{student} scored {score} in {subject}")
````

# Python List Methods

| Method         | Description                                  | Example                                 |
|----------------|----------------------------------------------|-----------------------------------------|
| `append()`     | Adds an item to the end of the list          | `fruits.append("apple")`                |
| `insert()`     | Inserts item at a specific index              | `fruits.insert(1, "orange")`            |
| `remove()`     | Removes the first matching value              | `fruits.remove("banana")`               |

| `clear()`      | Removes all elements from the list            | `fruits.clear()`                        |
| `index()`      | Returns the index of the first matching item  | `fruits.index("apple")`                 |
| `count()`      | Counts how many times a value appears         | `fruits.count("apple")`                 |
| `sort()`       | Sorts the list in ascending order (in-place)  | `numbers.sort()`                        |
| `reverse()`    | Reverses the list (in-place)                  | `fruits.reverse()`                      |
| `copy()`       | Returns a shallow copy of the list            | `new_list = fruits.copy()`              |
| `extend()`     | Adds elements from another list               | `list1.extend(list2)`                   |

## Step1: Employee Dictionary

````python
## employee performance scores

employees = {
    "Aisha": 92,
    "Rahul": 89,
    "Meera": 67,
    "Vijay": "N/A",
    "Meena": 78
}
````

## Step 2: Function to Evaluate Performance

````python
def evaluate_score(score):
  """
  Returns performance category based on score.
  Raises ValueError if score is not a valid number.
  """
  #try:
  score = float(score)
  if score >= 85:
    return "Excellent"
  elif score >= 70:
    return "Good"
  else:
    return "Needs Improvement"

  # except ValueError:
  #   raise ValueError("Invalid Score: Must be a number")
````

## Step 3: Lambda Function for Message Formatting

````python
format_msg = lambda name, rating: f"{name} -> {rating} \n"
````

 ## Step 4: Generate and Save Report

````python
def generate_report(employees, output_file = "performance_report.txt"):
  with open(output_file, "w") as file:
    file.write("Employee performance report \n")
    file.write("=" * 35 + "\n")

    for name, score in employees.items():
      try:
        rating = evaluate_score(score)
        line = format_msg(name, rating)

      except ValueError as e:
        line = format_msg(name, "Invalid Score")
      file.write(line)

    print(f"Report Successfully written to '{output_file}")
````

| **B. Cap/Floor (Winsorization)** | When you dont want to delete data but limit impact |
| **C. Replace with Median/Mean**  | For small datasets with minimal skew                |
| **D. Use Transformation**        | To reduce the effect (log, sqrt, etc.)              |
| **E. Binning/Clipping**          | To group values or limit extremes                   |
| **F. Treat case-by-case**        | If outliers carry meaning (like fraud detection)    |


#### Basic Outlier Treatment - using the IQR method to cap/ clip/ floor outliers: - this technique is called as __winsorization__

````python
def cap_outliers(series):
  Q1 = series.quantile(0.25)
  Q3 = series.quantile(0.75)
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  return series.clip(lower, upper)
````

````python
for col in numeric_cols:
  df[col] = cap_outliers(df[col])
````

**Regular Expressions** are special patterns used to match __text__.

Why __Regex__?

- **Find** specific patterns in text (emails, ph numbers, names)

| Pattern | Meaning                            | Example Match                    |
| ------- | ---------------------------------- | -------------------------------- |
| `.`     | Any character except newline       | `a.c`  `abc`, `a7c`             |
| `\d`    | Digit (09)                        | `\d\d`  `23`, `99`              |
| `\w`    | Word character (a-z, A-Z, 0-9, \_) | `\w+`  `Hello123`               |
| `\s`    | Whitespace (space, tab, etc.)      | `\s+`  `'   '`                  |
| `^`     | Start of string                    | `^Hello` matches `Hello World`   |

| `[]`    | Set of characters                  | `[aeiou]`  vowels               |
| `+`     | One or more                        | `a+`  `a`, `aaa`                |
| `*`     | Zero or more                       | `a*`  \`\`, `aaa`               |
| `?`     | Zero or one                        | `colou?r`  `color` or `colour`  |


| Function       | Purpose                                           |

| `re.sub()`     | **Substitute/replace** matching patterns          |

````python
import re

time_text = "The meeting is at 14:30."

# Pattern to match time in HH:MM format (24-hour)
# This regex pattern matches a time in 24-hour HH:MM format:
# - \b: Word boundary to ensure we match whole time strings
# - ([01]\d|2[0-3]): Matches hours from 00 to 23
#     - [01]\d: Matches 00-19
#     - 2[0-3]: Matches 20-23
# - : : Literal colon separating hours and minutes
# - [0-5]\d: Matches minutes from 00 to 59
# - \b: Word boundary to end the match
time_pattern = r'\b([01]\d|2[0-3]):[0-5]\d\b'

time_match = re.search(time_pattern, time_text)
if time_match:
    print("Matched time:", time_match.group())
else:
    print("No time found.")
````
