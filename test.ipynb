```python
# Example 1: Create a list of squares from 0 to 9
squares = [x**2 for x in range(10)]
print("Squares:", squares)

# Example 2: Filter even numbers from a list
evens = [num for num in range(20) if num % 2 == 0]
print("Even numbers:", evens)

# Example 3: Convert a list of strings to uppercase
words = ['apple', 'banana', 'cherry']
upper_words = [word.upper() for word in words]
print("Uppercase words:", upper_words)

# Example 4: Create a list of tuples (number, square)
tuples = [(x, x**2) for x in range(5)]
print("Tuples (number, square):", tuples)

# Example 5: Flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [item for row in matrix for item in row]
print("Flattened matrix:", flattened)
```
